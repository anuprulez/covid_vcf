```python
import pandas as pd
import numpy as np
```

```python
boston_all_path = "../data/outputs/Boston/all/"
pre_cog_uk_all_path = "../data/outputs/Pre-COG-UK/all/"
post_cog_uk_all_path = "../data/outputs/Post-COG-UK/all/"
```

```python
boston_merged_cluster_path = boston_all_path + "merged_clusters_of_mutations.csv"
pre_coguk_merged_cluster_path = pre_cog_uk_all_path + "merged_clusters_of_mutations.csv"
post_coguk_merged_cluster_path = post_cog_uk_all_path + "merged_clusters_of_mutations.csv"

boston_same_pos_path = boston_all_path + "same_pos_diff_ref_alt.csv"
pre_coguk_same_pos_path = pre_cog_uk_all_path + "same_pos_diff_ref_alt.csv"
post_coguk_same_pos_path = post_cog_uk_all_path + "same_pos_diff_ref_alt.csv"
```

```python
df_boston_mc = pd.read_csv(boston_merged_cluster_path)
df_boston_mc = df_boston_mc.sort_values(by=["POS"])
df_boston_mc.to_csv(boston_merged_cluster_path)
```

```python
df_pre_coguk_mc = pd.read_csv(pre_coguk_merged_cluster_path)
df_pre_coguk_mc = df_pre_coguk_mc.sort_values(by=["POS"])
df_pre_coguk_mc.to_csv(pre_coguk_merged_cluster_path)
```

```python
df_post_coguk_mc = pd.read_csv(post_coguk_merged_cluster_path)
df_post_coguk_mc = df_post_coguk_mc.sort_values(by=["POS"])
df_post_coguk_mc.to_csv(post_coguk_merged_cluster_path)
```

```python
df_boston_spos = pd.read_csv(boston_same_pos_path)
df_pre_coguk_spos = pd.read_csv(pre_coguk_same_pos_path)
df_post_coguk_spos = pd.read_csv(post_coguk_same_pos_path)
```

```python
def find_common_mutations_mc():
    max_pos = 30000
    common_df = None
    for i in range(max_pos):
        try:        
            POS = str(i)
            boston_rows = df_boston_mc[df_boston_mc["POS"] == i]
            pre_rows = df_pre_coguk_mc[df_pre_coguk_mc["POS"] == i]
            post_rows = df_post_coguk_mc[df_post_coguk_mc["POS"] == i]
            
            if len(boston_rows) > 0 and len(pre_rows) > 0 and len(post_rows) > 0: 
                boston_rows["Dataset_Name"] = np.repeat("Boston", len(boston_rows))
                pre_rows["Dataset_Name"] = np.repeat("PRE-COG-UK", len(pre_rows))
                post_rows["Dataset_Name"] = np.repeat("POST-COG-UK", len(post_rows))
                
                if common_df is None:
                    common_df = pd.concat([boston_rows, pre_rows], ignore_index=True)
                    common_df = pd.concat([common_df, post_rows], ignore_index=True)
                else:
                    common_df = pd.concat([common_df, boston_rows], ignore_index=True)
                    common_df = pd.concat([common_df, pre_rows], ignore_index=True)
                    common_df = pd.concat([common_df, post_rows], ignore_index=True)
        except Exception:
            print("POS {} not present in all three datasets".format(str(i)))
    common_df = common_df.drop(["SampleNames"], axis=1)
    common_df.to_csv("../data/outputs/Common/all/common_pos_mc_boston_pre_post_cog_uk.csv", index=False)
```

```python
find_common_mutations_mc()
```

```python
def find_common_mutations_mc_pre_post():
    max_pos = 30000
    common_df = None
    for i in range(max_pos):
        try:        
            POS = str(i)
            pre_rows = df_pre_coguk_mc[df_pre_coguk_mc["POS"] == i]
            post_rows = df_post_coguk_mc[df_post_coguk_mc["POS"] == i]
            
            if len(pre_rows) > 0 and len(post_rows) > 0: 
                pre_rows["Dataset_Name"] = np.repeat("PRE-COG-UK", len(pre_rows))
                post_rows["Dataset_Name"] = np.repeat("POST-COG-UK", len(post_rows))
                
                if common_df is None:
                    common_df = pd.concat([pre_rows, post_rows], ignore_index=True)
                else:
                    common_df = pd.concat([common_df, pre_rows], ignore_index=True)
                    common_df = pd.concat([common_df, post_rows], ignore_index=True)
        except Exception:
            print("POS {} not present in pre and post cog uk datasets".format(str(i)))
    common_df = common_df.drop(["SampleNames"], axis=1)
    common_df.to_csv("../data/outputs/Common/all/common_pos_mc_pre_post_cog_uk.csv", index=False)
```
